```python
order = ['date', 'opponent', 'goals', 'assists', 'points', 'plusMinus', 'powerPlayGoals', 'powerPlayPoints','powerPlayTimeOnIce','shortHandedGoals', 'gameWinningGoals',
       'overTimeGoals', 'shots', 'penaltyMinutes', 'shifts', 'timeOnIce', 'isHome', 'isWin', 'isOT', 'season']
```

```python
def get_game_info(game):
    df = game['stat']
    df = pd.DataFrame.from_dict(df, orient='index')
    game_id = game['game']['gamePk']
    df = df.T
    df.index = [game_id]
    df['date'] = game['date']
    df['opponent'] = game['opponent']['name']
    df['isHome'] = game['isHome']
    df['isWin'] = game['isWin']
    df['isOT'] = game['isOT']
    df['season'] = game['season']
    df['date']=pd.to_datetime(df['date'], format='%Y-%m-%d')
    df['timeOnIce']=pd.to_datetime(df['timeOnIce'], format='%M:%S')
    df['powerPlayTimeOnIce']=pd.to_datetime(df['powerPlayTimeOnIce'], format='%M:%S')
    df = df[order]
    return df
```

```python
get_game_info(games16_17[0])
```

# Пытаемся собрать все игры, все сезоны (на момент создания файла)

```python
start_hnl = '2005'
base_URL = 'https://statsapi.web.nhl.com/api/v1/people/8471214/stats/?stats=gameLog&season='
seasons = []
while start_hnl!='2023':
    season = start_hnl + str(int(start_hnl)+1)
    seasons.append(season)
    start_hnl = str(int(start_hnl)+1)
print(seasons)    
```

Формируем список сезонов, чтобы автоматически подставить в ссылку и затем в функцию

```python
total_df = pd.DataFrame()
for i in range(len(seasons)):
    url = 'https://statsapi.web.nhl.com/api/v1/people/8471214/stats/?stats=gameLog&season='+seasons[i]
    stats = requests.get(url=url).json()
    games = stats['stats'][0]['splits']
    for i in range(len(games)):
        df = get_game_info(games[i])
        total_df = pd.concat([total_df,df], axis=0)
total_df
```

```python
total_df.info()
```

```python
total_df = total_df.sort_values(by=['date'])
```

```python
total_df[total_df['season']=='20162017']
```
