```python
model_base_1.shots = model_base_1.shots.astype(int)
model_base_1.shifts = model_base_1.shifts.astype(int)
model_base_1.penaltyMinutes = model_base_1.penaltyMinutes.astype(int)
```

```python
time_check = model_base_1.timeOnIce.astype(str)
time_check
```

```python
model_base_1.timeOnIce = model_base_1.timeOnIce.apply(lambda x: ((datetime.strptime(x, '%M:%S')-datetime(1900,1,1)).total_seconds()))
```

```python
model_base_1.powerPlayTimeOnIce = model_base_1.powerPlayTimeOnIce.apply(lambda x: ((datetime.strptime(x, '%M:%S')-datetime(1900,1,1)).total_seconds()))
```

```python
ovi_bd = datetime(1985,9,17)
model_base_1.date = model_base_1.date.apply(lambda x: (x-ovi_bd).days)
```

```python
model_base_1.isHome = model_base_1.isHome.astype(int)
model_base_1.isOT = model_base_1.isOT.astype(int)
model_base_1.isWin = model_base_1.isWin.astype(int)
```
